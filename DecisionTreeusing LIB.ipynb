import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('enjoysport.csv')
dataset = dataset.drop(['Day'], axis=1)

# Encode categorical variables into numeric form
label_encoder = LabelEncoder()
for column in dataset.columns:
    dataset[column] = label_encoder.fit_transform(dataset[column])

# Separate features and target
features = dataset.drop('play', axis=1)
target = dataset['play']

# Train the decision tree model
decision_tree = DecisionTreeClassifier(criterion='entropy')
decision_tree.fit(features, target)

# Visualize the decision tree
plt.figure(figsize=(12, 8))
tree.plot_tree(decision_tree, feature_names=features.columns, class_names=['no', 'yes'], filled=True)
plt.show()


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
dataset = pd.read_csv('enjoysport.csv')
dataset = dataset.drop(['Day'], axis=1)
label_encoder = LabelEncoder()
for column in dataset.columns:
dataset[column] = label_encoder.fit_transform(dataset[column])
features = dataset.drop('play', axis=1)
target = dataset['play']
decision_tree = DecisionTreeClassifier(criterion='entropy')
decision_tree.fit(features, target)
plt.figure(figsize=(12, 8))
tree.plot_tree(decision_tree, feature_names=features.columns,
class_names=['no', 'yes'], filled=True)
plt.show();
```
