```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
iris_dataset = pd.read_csv("iris.csv")
iris_dataset.drop('Id', axis=1, inplace=True)
feature_columns = iris_dataset.drop('Species', axis=1)
species_labels = iris_dataset['Species']
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(feature_columns)
normalized_features_df = pd.DataFrame(normalized_features,
columns=['scaled_SepalLength', 'scaled_SepalWidth',
'scaled_PetalLength', 'scaled_PetalWidth'])
scaled_iris_data = pd.concat([normalized_features_df, species_labels],
axis=1)
pca_result = PCA(n_components=2).fit_transform(feature_columns)
pca_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
pca_visualization_df = pd.concat([pca_df, species_labels], axis=1)
X_train, X_test, y_train, y_test = train_test_split(feature_columns,
species_labels, test_size=0.3, random_state=0)
svm_model = SVC(kernel='linear').fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix_df = pd.DataFrame(conf_matrix,
index=['setosa', 'versicolor',
'virginica'],
columns=['setosa', 'versicolor',
'virginica'])
plt.figure(figsize=(5.5, 4))
sns.heatmap(conf_matrix_df, annot=True, cmap="Blues", fmt='g')
plt.title('SVM Linear Kernel')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
